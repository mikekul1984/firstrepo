```python
# First let's import the dataset, using Pandas.
import pandas as pd

fileName = "/Users/alexwoods/Desktop/IRIS_dataset/iris.txt"
iris = pd.read_csv(fileName, header = None)
iris.head()


# sepal length in cm
#    2. sepal width in cm
#    3. petal length in cm
#    4. petal width in cm
#    5. class: 

iris.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
print(iris)
```

```python
# Now, let's split the data into training and test data. We need to assure
# that an equal amount of each class is in each. 

iris['class2'] = iris['class']
grouped = iris.groupby('class2')
grouped.first()        # shows the first row of each class
a, b, c = [grouped.get_group(x) for x in grouped.groups]        # splits into subsets

import random

# to randomly sample, presumably, there is some order in the data.
rows = random.sample(a.index.tolist(), 35)        
trainA = a.ix[rows]
testA = a.drop(rows)

rows = random.sample(b.index.tolist(), 35)  
trainB = b.ix[rows]
testB = b.drop(rows)

rows = random.sample(c.index.tolist(), 35)
trainC = c.ix[rows]
testC = c.drop(rows)

# now we have to combine them
train = trainA.append(trainB)
train = train.append(trainC)

test = testA.append(testB)
test = test.append(testC)
```

```python
# Ok, now we have good training and test data! That took forever.. Might write a seperate blog post
# just on that haha

from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100)    # 100 decision trees is a good enough number

# however, are data has to be in a numpy array in order for the random forest algorithm to except it!
cols = ['petal_length', 'petal_width', 'sepal_length', 'sepal_width']
colsRes = ['class']
trainArr = train.as_matrix(cols)    # train array
trainRes = train.as_matrix(colsRes) # train results

rf.fit(trainArr, trainRes)          # finally, we fit the data to the algorithm!!! :)

# note - you might get an warning saying you entered a 2 column vector..ignore it.
```

```python
# put the test results in the same format!
testArr = test.as_matrix(cols)

results = rf.predict(testArr)

# something I like to do is to add it back to the dataframe, so I can compare side-by-side
test['predictions'] = results
test
```
